```python
import os 
import pandas as pd
from shutil import copyfile, copy2
```

```python
ZP_nummer = 9
```

```python
ZP_csv = pd.read_csv('/home/jan/Documents/Trainingsdaten/ZPs/ZP{}/grid_size=874_threshold_porosity=125.csv'.format(ZP_nummer))
array_path = '/home/jan/Documents/Trainingsdaten/ZPs/ZP{}/1layer/Grid_size_874/'.format(ZP_nummer)
folder_pores = '/home/jan/Documents/Trainingsdaten/datasets/Grid_size_874_binary/porosity/'
folder_no_pores = '/home/jan/Documents/Trainingsdaten/datasets/Grid_size_874_binary/no_porosity/'
```

```python
ZP_csv
```

```python
for index,row in ZP_csv.iterrows():
    num_slice = row['Slice']
    x = row['x-grid']
    y = row['y-grid']
    Poren = row['Poren']
    src = array_path + 'Slice' + str("{:05d}".format(num_slice)) + '_x:' + str(x) + '_y:' + str(y) + '.npy'
    if os.path.isfile(src):
        if Poren == 0:
            dst = folder_no_pores + 'ZP{}_'.format(ZP_nummer) + 'Slice' + str("{:05d}".format(num_slice)) + '_x:' + str(x) + '_y:' + str(y) + '.npy'
                
        elif Poren == 1: 
            dst = folder_pores + 'ZP{}_'.format(ZP_nummer) + 'Slice' + str("{:05d}".format(num_slice)) + '_x:' + str(x) + '_y:' + str(y) + '.npy'
            
        copyfile(src, dst)             
```
