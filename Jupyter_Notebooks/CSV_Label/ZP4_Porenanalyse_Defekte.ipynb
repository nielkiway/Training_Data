```python
import pandas as pd
import numpy as np
import math
import itertools
```

```python
data = pd.read_csv('/home/jan/Documents/Trainingsdaten/CT-Analyse_QualiPro_ZP_4/Porenanalyse_Defekte_ZP4_17.12.2019.csv', delimiter =';')
```

```python
data
```

```python
data_reduced = data[['Wahrscheinlichkeit', 'Radius [mm]', 'Durchmesser [mm]', 'Mitte x [mm]','Mitte y [mm]','Mitte z [mm]','Projizierte Größe x [mm]','Projizierte Größe y [mm]','Projizierte Größe z [mm]']]
```

```python
data_reduced = data_reduced.stack().str.replace(',','.').unstack()
```

```python
#the two different coordinate systems need to be transferred -> Adding half the diameter to all the Mitte x,y values 
diameter = 6.92359 #mm from solidworks
radius = diameter/2
```

```python
mitte_x = np.asarray(data_reduced['Mitte x [mm]'], dtype = float)+radius #- data_reduced['Projizierte Größe x [mm]']/2
proj_x = np.asarray(data_reduced['Projizierte Größe x [mm]'], dtype = float)

mitte_y = np.asarray(data_reduced['Mitte y [mm]'], dtype = float)+radius #- data_reduced['Projizierte Größe x [mm]']/2
proj_y = np.asarray(data_reduced['Projizierte Größe y [mm]'], dtype = float)

mitte_z = np.asarray(data_reduced['Mitte z [mm]'], dtype = float) #- data_reduced['Projizierte Größe x [mm]']/2
proj_z = np.asarray(data_reduced['Projizierte Größe z [mm]'], dtype = float)
```

```python
mitte_x
```

```python
data_minX = (mitte_x-proj_x/2).round(4)
data_maxX = (mitte_x+proj_x/2).round(4)

data_minY = (mitte_y-proj_y/2).round(4)
data_maxY = (mitte_y+proj_y/2).round(4)

data_minZ = (mitte_z-proj_z/2).round(4)
data_maxZ = (mitte_z+proj_z/2).round(4)

data_reduced["Pore min x"] = data_minX
data_reduced["Pore max x"] = data_maxX

data_reduced["Pore min y"] = data_minY
data_reduced["Pore max y"] = data_maxY

data_reduced["Pore min z"] = data_minZ
data_reduced["Pore max z"] = data_maxZ

data_reduced['Mitte x [mm]'] = mitte_x
data_reduced['Mitte y [mm]'] = mitte_y
```

```python
data_reduced
```

```python
example_data = data_reduced[data_reduced['Durchmesser [mm]'].astype(float) > 1.5]
```

```python
example_data
```

```python
#Berechnung der betroffenenen Voxel 
#Annahme Voxelkantenlänge 500 µm 
#vorerst nur x,y berücksichtigt
layer_height = 0.05
voxel_len = 0.5
vox_layer = 3 #slices pro vox_layer
x_wert = 1.6
x_wert/voxel_len 
math.floor(x_wert/voxel_len)-1 #-1 wegen 0er Voxeln 
```

```python
#asd = ((np.asarray(example_data['Pore min x'], dtype = float)/voxel_len-1))
min_vox_x = np.floor((np.asarray(example_data['Pore min x'], dtype = float)/voxel_len-1)).astype(int)
max_vox_x = np.floor((np.asarray(example_data['Pore max x'], dtype = float)/voxel_len-1)).astype(int)
min_vox_y = np.floor((np.asarray(example_data['Pore min y'], dtype = float)/voxel_len-1)).astype(int)
max_vox_y = np.floor((np.asarray(example_data['Pore max y'], dtype = float)/voxel_len-1)).astype(int)
min_vox_z = np.floor((np.asarray(example_data['Pore min z'], dtype = float)/layer_height-1)).astype(int)
max_vox_z = np.floor((np.asarray(example_data['Pore max z'], dtype = float)/layer_height-1)).astype(int)
```

```python
example_data['min vox x'] = min_vox_x
example_data['max vox x'] = max_vox_x
example_data['min vox y'] = min_vox_y
example_data['max vox y'] = max_vox_y
example_data['min slice z'] = min_vox_y
example_data['max slice z'] = max_vox_y
```

```python
#getting all x,y combinations for a pore 
list_x = [i for i in range(example_data.iloc[1]['min vox x'],example_data.iloc[1]['max vox x']+1)]
list_y = [i for i in range(example_data.iloc[1]['min vox y'],example_data.iloc[1]['max vox y']+1)]
list_z = [i for i in range(example_data.iloc[1]['min slice z'],example_data.iloc[1]['max slice z']+1)]
```
