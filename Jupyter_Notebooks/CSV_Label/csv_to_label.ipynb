```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/home/jan/Documents/Trainingsdaten/Test__19017003_ZP1.csv', delimiter =';')
```

```python
data
```

```python
data_reduced = data[['Wahrscheinlichkeit', 'Radius [mm]', 'Durchmesser [mm]', 'Mitte x [mm]','Mitte y [mm]','Mitte z [mm]','Pos. x [mm]','Pos. y [mm]','Pos. z [mm]','Projizierte Größe x [mm]','Projizierte Größe y [mm]','Projizierte Größe z [mm]']]
```

```python
data_reduced.head()
```

```python
data_reduced = data_reduced.stack().str.replace(',','.').unstack()
```

```python
data_reduced
```

```python
mitte_x = np.asarray(data_reduced['Mitte x [mm]'], dtype = float) #- data_reduced['Projizierte Größe x [mm]']/2
proj_x = np.asarray(data_reduced['Projizierte Größe x [mm]'], dtype = float)

mitte_y = np.asarray(data_reduced['Mitte y [mm]'], dtype = float) #- data_reduced['Projizierte Größe x [mm]']/2
proj_y = np.asarray(data_reduced['Projizierte Größe y [mm]'], dtype = float)

mitte_z = np.asarray(data_reduced['Mitte z [mm]'], dtype = float) #- data_reduced['Projizierte Größe x [mm]']/2
proj_z = np.asarray(data_reduced['Projizierte Größe z [mm]'], dtype = float)
```

```python
data_minX = (mitte_x-proj_x/2).round(2)
data_maxX = (mitte_x+proj_x/2).round(2)

data_minY = (mitte_y-proj_y/2).round(2)
data_maxY = (mitte_y+proj_y/2).round(2)

data_minZ = (mitte_z-proj_z/2).round(2)
data_maxZ = (mitte_z+proj_z/2).round(2)
```

```python
data_reduced["Pore min x"] = data_minX
data_reduced["Pore max x"] = data_maxX

data_reduced["Pore min y"] = data_minY
data_reduced["Pore max y"] = data_maxY

data_reduced["Pore min z"] = data_minZ
data_reduced["Pore max z"] = data_maxZ
```
